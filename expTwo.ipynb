```python
import hashlib
```

```python
def hash_data(data):
    if isinstance(data, str):
        data = data.encode('utf-8')

    sha256_hash = hashlib.sha256()
    sha256_hash.update(data)

    return sha256_hash.hexdigest()
```

```python
original_dataOne = "This is some data to protect"
hashed_dataOne = hash_data(original_dataOne)

print("Original Data:", original_dataOne)
print("Hash:", hashed_dataOne)
print("Hash length: ", len(hashed_dataOne))
```

```python
original_dataTwo = "Shriswarup Sawant"
hashed_dataTwo = hash_data(original_dataTwo)

print("Original Data:", original_dataTwo)
print("Hash:", hashed_dataTwo)
print("Hash length: ", len(hashed_dataTwo))
```

```python
if ((hashed_dataOne != hashed_dataTwo) and (len(original_dataOne) != len(original_dataTwo)) and (original_dataOne != original_dataTwo)):
    print("Both data values are different with different data length and hashes are diffenent but with same length.")
```
