<h1><b>Proof Of Work<b></h1>

```python
import hashlib
import time
```

```python
def proof_of_work(difficulty):
    prefix = '0' * difficulty
    nonce = 0
    start_time = time.time()
    while True:
        text = f'{nonce}'.encode()
        hash_result = hashlib.sha256(text).hexdigest()
        if hash_result.startswith(prefix):
            end_time = time.time()
            print(f"Found hash: {hash_result}")
            print(f"Nonce: {nonce}")
            print(f"Time taken: {end_time - start_time:.4f} seconds")
            break
        nonce += 1
```

```python
# Try it with a difficulty of 4
proof_of_work(difficulty=4)
```

<h1><b>Proof Of Stake<b></h1>

```python
import random
```

```python
# Simulated validator data: (validator name, stake amount)
validators = {
    "Alice": 50,
    "Bob": 30,
    "Charlie": 20
}
```

```python
def proof_of_stake(validators):
    total_stake = sum(validators.values())
    pick = random.uniform(0, total_stake)
    current = 0

    for validator, stake in validators.items():
        current += stake
        if current >= pick:
            print(f"Selected validator: {validator}")
            break
```

```python
# Simulate 5 rounds of validator selection
for i in range(5):
    print(f"\n  Round {i+1}")
    proof_of_stake(validators)
```
